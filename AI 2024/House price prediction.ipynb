# House Price Prediction

```python
#import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
#lấy dữ liệu
data=pd.read_csv("train.csv")
data.head()
```

```python
data.columns
```

```python
#Lấy những tính năng, đặc trưng muốn dùng để định giá căn nhà
features = ['OverallQual', 'OverallCond','LotArea','YearBuilt','1stFlrSF', '2ndFlrSF','FullBath','BedroomAbvGr','GarageArea']
```

```python
X = data[features]
X.head()
```

```python
y = data['SalePrice']
y.head()
```

```python
#Phân loại tập train và tập test
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.8, test_size=0.2, random_state=0)
print('X_train shape:', X_train.shape, '\nX_test shape:', X_test.shape)
print('y_train shape:', y_train.shape, '\ny_test shape:', y_test.shape)
#train size = 80%
#test size = 20%
```

```python
#Chuẩn hoá dữ liệu
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test= sc.transform(X_test)
```

```python
#tạo model
from sklearn.tree import DecisionTreeRegressor
model= DecisionTreeRegressor(max_depth=10, min_samples_leaf= 4, min_samples_split= 5)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
#dự đoán
```

```python
pd.DataFrame({'y': y_test, 'y_predict':y_pred})
#so sánh, đối chiếu
```
